**自助法**（Bootstrap）是机器学习中一种重要的重采样技术，通过**有放回抽样**来估计统计量的分布。

## 实验流程
1. 生成 $N$ 个服从标准正态分布 $\mathcal{N}(0,1)$ 的随机数
2. 从这 $N$ 个数中，用自助法（有放回）抽样 $M$ 次，每次抽取 $N$ 个数，共得到 $M$ 个**自助样本集**
3. 计算每个自助样本集的均值和方差，并分析其分布

## 第一步：导入库并设置参数

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from scipy import stats

# 设置随机种子，保证结果可复现
np.random.seed(42)

# 参数设置
N = 10    # 原始样本数量
M = 1000  # 自助抽样次数

print(f"参数设置：N = {N}（原始样本量），M = {M}（自助抽样次数）")
```

## 第二步：生成 N 个标准正态随机数

```python
# 生成 N= 个服从标准正态分布 N(0,1) 的随机数
original_data = np.random.randn(N)

print(f"原始样本数据（N={N}）：")
print("-" * 45)
for i, val in enumerate(original_data):
    print(f"  x[{i+1:2d}] = {val:+.4f}")
print("-" * 45)
print(f"  样本均值  : {original_data.mean():.4f}")
print(f"  样本方差  : {original_data.var():.4f}")
print(f"  样本标准差: {original_data.std():.4f}")
```

## 第三步：执行自助法（有放回抽样）M 次

```python
# 存储每次自助样本的均值和方差
bootstrap_means = np.zeros(M)
bootstrap_vars  = np.zeros(M)

# 自助抽样：每次从原始数据中有放回地抽取 N 个样本
for i in range(M):
    # 有放回抽样（replace=True）
    bootstrap_sample = np.random.choice(original_data, size=N, replace=True)
    bootstrap_means[i] = bootstrap_sample.mean()
    bootstrap_vars[i]  = bootstrap_sample.var()

print(f"自助法抽样完成！共生成 {M} 个自助样本集")
```

## 第四步：计算并汇总统计结果

```python
# 计算 M 个自助样本均值和方差的统计量
mean_of_means = bootstrap_means.mean()
var_of_means  = bootstrap_means.var()
std_of_means  = bootstrap_means.std()

mean_of_vars  = bootstrap_vars.mean()
var_of_vars   = bootstrap_vars.var()
std_of_vars   = bootstrap_vars.std()

print("=" * 50)
print("         自助法统计结果汇总")
print("=" * 50)
print()
print(f"【原始样本（N={N}）】")
print(f"  均值 = {original_data.mean():.6f}")
print(f"  方差 = {original_data.var():.6f}")
print()
print(f"【M={M} 个自助样本均值的分布】")
print(f"  均值的均值 = {mean_of_means:.6f}")
print(f"  均值的方差 = {var_of_means:.6f}")
print(f"  均值的标准差 = {std_of_means:.6f}")
print()
print(f"【M={M} 个自助样本方差的分布】")
print(f"  方差的均值 = {mean_of_vars:.6f}")
print(f"  方差的方差 = {var_of_vars:.6f}")
print(f"  方差的标准差 = {std_of_vars:.6f}")
print()
print("【理论参考值（标准正态分布）】")
print(f"  总体均值 = 0.000000")
print(f"  总体方差 = 1.000000")
print(f"  均值标准误（理论）= 1/sqrt(N) = {1/np.sqrt(N):.6f}")
print("=" * 50)
```

## 第五步：可视化分析

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle(f'Bootstrap Method Experiment  (N={N}, M={M})', 
             fontsize=15, fontweight='bold', y=1.01)

colors = {'hist': '#4C72B0', 'kde': '#DD8452', 'line': '#C44E52', 'orig': '#2CA02C'}

# ── 子图1：原始样本可视化 ──────────────────────────────
ax1 = axes[0, 0]
x_range = np.linspace(-4, 4, 300)
ax1.plot(x_range, stats.norm.pdf(x_range), color=colors['line'],
         lw=2, label='Standard Normal N(0,1)')
ax1.scatter(original_data, np.zeros(N), color=colors['orig'],
            s=100, zorder=5, label=f'Original N={N} samples')
for v in original_data:
    ax1.axvline(v, color=colors['orig'], alpha=0.3, linewidth=1)
ax1.axvline(original_data.mean(), color='purple', lw=2, linestyle='--',
            label=f'Sample mean = {original_data.mean():.3f}')
ax1.set_title(f'Step 1: Original Sample (N={N})', fontsize=12, fontweight='bold')
ax1.set_xlabel('Value'); ax1.set_ylabel('Density')
ax1.legend(fontsize=8); ax1.grid(alpha=0.3)

# ── 子图2：Bootstrap 均值分布 ─────────────────────────
ax2 = axes[0, 1]
ax2.hist(bootstrap_means, bins=40, density=True,
         color=colors['hist'], alpha=0.7, edgecolor='white', label='Bootstrap means')
# 拟合正态曲线
mu_fit, std_fit = stats.norm.fit(bootstrap_means)
x_fit = np.linspace(bootstrap_means.min(), bootstrap_means.max(), 200)
ax2.plot(x_fit, stats.norm.pdf(x_fit, mu_fit, std_fit),
         color=colors['kde'], lw=2.5, label=f'Fitted Normal\nμ={mu_fit:.3f}, σ={std_fit:.3f}')
ax2.axvline(mean_of_means, color=colors['line'], lw=2, linestyle='--',
            label=f'Mean of means = {mean_of_means:.3f}')
ax2.axvline(0, color='gray', lw=1.5, linestyle=':', label='True mean = 0')
ax2.set_title(f'Bootstrap Means Distribution (M={N})', fontsize=12, fontweight='bold')
ax2.set_xlabel('Bootstrap Sample Mean'); ax2.set_ylabel('Density')
ax2.legend(fontsize=8); ax2.grid(alpha=0.3)

# ── 子图3：Bootstrap 方差分布 ─────────────────────────
ax3 = axes[1, 0]
ax3.hist(bootstrap_vars, bins=40, density=True,
         color='#55A868', alpha=0.7, edgecolor='white', label='Bootstrap variances')
mu_v, std_v = stats.norm.fit(bootstrap_vars)
x_v = np.linspace(bootstrap_vars.min(), bootstrap_vars.max(), 200)
ax3.plot(x_v, stats.norm.pdf(x_v, mu_v, std_v),
         color=colors['kde'], lw=2.5, label=f'Fitted Normal\nμ={mu_v:.3f}, σ={std_v:.3f}')
ax3.axvline(mean_of_vars, color=colors['line'], lw=2, linestyle='--',
            label=f'Mean of vars = {mean_of_vars:.3f}')
ax3.axvline(1.0, color='gray', lw=1.5, linestyle=':', label='True variance = 1')
ax3.set_title(f'Bootstrap Variances Distribution (M={M})', fontsize=12, fontweight='bold')
ax3.set_xlabel('Bootstrap Sample Variance'); ax3.set_ylabel('Density')
ax3.legend(fontsize=8); ax3.grid(alpha=0.3)

# ── 子图4：均值收敛曲线 ───────────────────────────────
ax4 = axes[1, 1]
cumulative_mean = np.cumsum(bootstrap_means) / np.arange(1, M+1)
cumulative_var  = np.cumsum(bootstrap_vars)  / np.arange(1, M+1)
ax4.plot(cumulative_mean, color=colors['hist'], lw=1.5, label='Cumulative mean of means')
ax4.plot(cumulative_var,  color='#55A868',      lw=1.5, label='Cumulative mean of variances')
ax4.axhline(0, color=colors['hist'], lw=1, linestyle='--', alpha=0.6, label='True mean = 0')
ax4.axhline(1, color='#55A868',      lw=1, linestyle='--', alpha=0.6, label='True variance = 1')
ax4.axhline(original_data.mean(), color='purple', lw=1, linestyle=':', alpha=0.8,
            label=f'Sample mean = {original_data.mean():.3f}')
ax4.set_title('Convergence of Bootstrap Estimates', fontsize=12, fontweight='bold')
ax4.set_xlabel('Number of Bootstrap Samples'); ax4.set_ylabel('Cumulative Estimate')
ax4.legend(fontsize=8); ax4.grid(alpha=0.3)
ax4.set_xlim(0, M)

plt.tight_layout()
plt.savefig(f'result/python/bootstrap_result_N={N}_M={M}.png', dpi=300, bbox_inches='tight')
plt.show()
print("图表已保存！")
```

## 第六步：置信区间估计

```python
# 利用自助法估计总体均值的 95% 置信区间（百分位法）
alpha = 0.05
ci_mean_low  = np.percentile(bootstrap_means, 100 * alpha / 2)
ci_mean_high = np.percentile(bootstrap_means, 100 * (1 - alpha / 2))

ci_var_low   = np.percentile(bootstrap_vars, 100 * alpha / 2)
ci_var_high  = np.percentile(bootstrap_vars, 100 * (1 - alpha / 2))

print("=" * 55)
print("     Bootstrap 95% 置信区间（百分位法）")
print("=" * 55)
print(f"  总体均值的置信区间: [{ci_mean_low:.4f}, {ci_mean_high:.4f}]")
print(f"  总体方差的置信区间: [{ci_var_low:.4f},  {ci_var_high:.4f}]")
print()
print("  真实总体均值 0 是否在均值置信区间内：",
      "✓ 是" if ci_mean_low <= 0 <= ci_mean_high else "✗ 否")
print("  真实总体方差 1 是否在方差置信区间内：",
      "✓ 是" if ci_var_low  <= 1 <= ci_var_high  else "✗ 否")
print("=" * 55)
```

# 第七步：验证课本P27公式(2.1)    

```python
# --- 解析计算 ---
prob_theory = (1 - 1/N) ** N
prob_limit  = np.exp(-1)

print('=' * 60)
print('某样本在一次自助抽样中始终不被选中的概率')
print('=' * 60)
print(f'  N            = {N}')
print(f'  (1-1/N)^N    = {prob_theory:.6f}')
print(f'  极限 e^(-1)  = {prob_limit:.6f}')
print(f'  两者差值     = {abs(prob_theory - prob_limit):.6f}')
print()

# --- 模拟实验 ---
np.random.seed(42)
never_seen_count = np.zeros(N, dtype=int)

for _ in range(M):
    indices = np.random.randint(0, N, size=N)
    sampled_set = set(indices)
    for j in range(N):
        if j not in sampled_set:
            never_seen_count[j] += 1

never_seen_prob = never_seen_count / M
empirical_mean  = never_seen_prob.mean()

print(f'  模拟实验结果（M={M} 次采样）：')
print(f"  {'样本索引':>6}  {'未被采到次数':>12}  {'实验概率':>10}")
print('  ' + '-' * 36)
for j in range(N):
    print(f'  {j:>6}  {never_seen_count[j]:>12}  {never_seen_prob[j]:>10.4f}')
print('  ' + '-' * 36)
print(f'  平均实验概率          {empirical_mean:>10.4f}')
print()
print(f'  理论值 (1-1/N)^N = {prob_theory:.4f}')
print(f'  极限值  e^(-1)   = {prob_limit:.4f}')
print(f'  实验均值         = {empirical_mean:.4f}')
print('=' * 60)
```

# 第八步：可视化极限

```python
# 可视化：(1-1/N)^N 随 N 增大收敛到 e^{-1}，以及各样本实验概率
N_values   = np.arange(1, 201)
prob_curve = (1 - 1/N_values) ** N_values

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f'OOB Probability Verification  (N={N}, M={M})', fontsize=13, fontweight='bold')

# 左图：收敛曲线
ax = axes[0]
ax.plot(N_values, prob_curve, color='#4C72B0', lw=2, label='$(1-1/N)^N$')
ax.axhline(prob_limit, color='#C44E52', lw=1.8, linestyle='--',
           label=f'Limit $e^{{-1}}$ = {prob_limit:.4f}')
ax.axvline(N, color='#2CA02C', lw=1.5, linestyle=':',
           label=f'Current N={N}, value={prob_theory:.4f}')
ax.scatter([N], [prob_theory], color='#2CA02C', s=80, zorder=5)
ax.set_title('$(1-1/N)^N$ Converges to $e^{-1}$', fontsize=11)
ax.set_xlabel('N (Sample Size)')
ax.set_ylabel('Probability')
ax.legend(fontsize=9)
ax.grid(alpha=0.3)
ax.set_ylim(0, 0.5)

# 右图：各样本的实验 OOB 概率
ax2 = axes[1]
ax2.bar(np.arange(N), never_seen_prob, color='#4C72B0', alpha=0.75,
        label='Empirical Probability (Never Selected)')
ax2.axhline(prob_theory, color='#DD8452', lw=2, linestyle='--',
            label=f'Theoretical Value $(1-1/N)^N$ = {prob_theory:.4f}')
ax2.axhline(prob_limit, color='#C44E52', lw=1.5, linestyle=':',
            label=f'Limit Value $e^{{-1}}$ = {prob_limit:.4f}')
ax2.axhline(empirical_mean, color='#2CA02C', lw=1.5, linestyle='-.',
            label=f'Empirical Mean = {empirical_mean:.4f}')
ax2.set_title(f'Empirical OOB Probability for Each Sample (N={N}, M={M})', fontsize=11)
ax2.set_xlabel('Sample Index')
ax2.set_ylabel('Proportion Never Selected')
ax2.set_xticks(np.arange(N))
ax2.legend(fontsize=9)
ax2.grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig(f'oob/python/bootstrap_oob_N={N}_M={M}.png', dpi=300, bbox_inches='tight')
plt.show()
print('OOB 验证图表已保存!')
```

## 实验总结

| 统计量 | 真实值（总体） | 原始样本估计 | Bootstrap 均值估计 |
|--------|:-----------:|:----------:|:-----------------:|
| **均值** | 0 | ≈ 样本均值 | ≈ 0（收敛） |
| **方差** | 1 | ≈ 样本方差 | ≈ 1（收敛） |
| **均值标准误** | $1/\sqrt{N}$ | — | Bootstrap标准差 |

**关键结论：**
- 自助法通过**有放回重采样**，可以从少量数据中估计统计量的**抽样分布**
- 随着抽样次数 $M$ 增大，Bootstrap 估计量**收敛**到真实值
- Bootstrap 均值分布近似**正态分布**，符合中心极限定理
- 可用 Bootstrap 分布的百分位数直接构造**置信区间**，无需假设数据分布
